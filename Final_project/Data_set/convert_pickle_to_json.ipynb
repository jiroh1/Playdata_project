```python
# import libraries
import pickle
import json
import sys
import os

# open pickle file
with open('insta_index.pickle', 'rb') as infile:
    obj = pickle.load(infile)

# convert pickle object to json object
json_obj = json.loads(json.dumps(obj, default=str))

#write the json file
with open(
        os.path.splitext('insta_index.pickle')[0] + '.json',
        'w',
        encoding='utf-8'
    ) as outfile:
    json.dump(json_obj, outfile, ensure_ascii=False, indent=4)
```
