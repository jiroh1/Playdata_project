```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리

df['시'] = df['지역'].str.split(' ').str[0]
df['구'] = df['지역'].str.split(' ').str[1]
df
```

```python
# 시 컬럼의 서울을 서울특별시로 변환

df['시']='서울특별시' 
df.head()
```

```python
# 기간 컬럼을 datetime 형식으로 변경

df['기간'] = pd.to_datetime(df['기간'], format='%Y%m%d')
df.head()

# astype 형식으로 변경하기

# df['기간']= df['기간'].astype('datetime64[ns]', format='%Y%m%d')

# # df.info()
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성

import datetime as dt

df['기간'] = pd.to_datetime(df['기간'])
df['년'] = df['기간'].dt.year
df['월'] = df['기간'].dt.month
df['일'] = df['기간'].dt.day
df['요일'] = df['기간'].dt.strftime('%A')
df['요일(number)'] = df['기간'].dt.strftime('%w')
df['주차'] = df['기간'].dt.strftime('%W')
df.head()
```

### 요일별 유종별 가격 변동

    pointplot 사용이유 : 기간의 연속성을 표현하기 위하여 선택

    ※ 고급휘발유는 토요일이 가장 가격이 높은 반면, 휘발유와 경유는 월, 화가 높았다.
    


```python
a = df.pivot_table(index=['요일(number)'], values=['휘발유', '고급휘발유', '경유']).sort_values('요일(number)')


plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.pointplot(x=a.index, y=a['고급휘발유'],color='lime')
plt.grid()

plt.xticks(rotation=45)
plt.subplot(1,3,2)
sns.pointplot(x=a.index, y=a['휘발유'],color="orangered")
plt.grid()

plt.xticks(rotation=45)
plt.subplot(1,3,3)
sns.pointplot(x=a.index, y=a['경유'])
plt.grid()

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

# 0은 일요일 ~ 6 토요일
# 결과 : 고급휘발유는 토요일이 가장 가격이 높은 반면, 휘발유와 경유는 월, 화가 높았다.
# 해석 : 대체적으로 수요일이 가장 저렴했다 , 수요일에 주유하는 것을 권장!
```


   * 결과 : 
    
    휘발유, 경유는 한 주가 시작되는 월 ~ 화에 기름값이 비싸고, 고급휘발유는 토요일이 상대적으로 비싼 경향을 보여준다.
    
    
    
   * 해석 :
    
    1. 고급휘발유는 주로 고급 차를 많이 사용하니, 주말에 많이 사용 , 휘발유, 경유는 비싸도 많이 써야해서 라고 오르지 않는가?
    2. 수요일이 대체적으로 기름값이 싸다. 수요일날 주유하는 것을 권장

### 주간별(연간) 유종별 가격 변동


1. 국제유가에 대한 생각이 없는 상태에서의 해석

        3월부터 국내 소비량의 증가로 값이 상승하는 것이 아닌가?

    

2. 국제유가가 국내유가에 반영되기까지 3주정도 걸린다.

        그 이유는 정유사에서 오일을 구매하기 까지의 2-3주정도의 term이 있기에 구내 유가에 반영되기 까지 3주정도가 생긴다.
        (단, 국내유가는 80%가 세금이기 때문에 국제유가 외에 변수가 있는 것은 배제하고 생각하였다.)
        
        
→ 국제 유가 관련 외부 데이터를 취합하면, 국내 유가가 국제 유가로부터 영향을 받아 변화하는 양상을 확인할 수 있을 것으로 보인다.



```python
c = df.pivot_table(index=['주차'], values=['고급휘발유', '휘발유', '경유']).sort_values('주차')
plt.figure(figsize=(14,20))
plt.subplot(3,1,1)
#sns.pointplot(x=a.index, y=a['고급휘발유'])
plt.plot(c.index,c['고급휘발유'], color='lime')
plt.title('고급휘발유')
plt.grid()
plt.xticks(rotation=45)
plt.yscale('linear')

plt.subplot(3,1,2)
#sns.pointplot(x=a.index, y=a['휘발유'])
plt.plot(c.index,c['휘발유'],color="orangered")
plt.title('휘발유')
plt.grid()
plt.xticks(rotation=45)


plt.subplot(3,1,3)
#sns.pointplot(x=a.index, y=a['경유'])
plt.plot(c.index,c['경유'])
plt.title('경유')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import folium
import json

geo_path = '../PJT/municipalities-geo-simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))


center = [37.541, 126.986]
m1 = folium.Map(location=center, zoom_start=10.5)
choropleth = folium.Choropleth(
        geo_data=geo_str,
        data=new_df,
        columns=('구','개수'),
        key_on='feature.properties.name', # name 매칭!
        fill_color='RdBu_r', # B < R
        legend_name='구별 주유소 개수').add_to(m1)
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'],labels=False)
)
m1


```
