```python
import urllib
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from pprint import pprint
import pandas as pd
```

```python
df = pd.DataFrame(columns = ["img","title","price"]) # 컬럼을 만든다.
df.index.name = "ID"
df
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
for page in range(1,2):
    url = f"https://store.musinsa.com/app/news/lists?year_date=2021&page={page}"
    driver.get(url)
    time.sleep(5)
    #pattern
    musinsa = driver.find_elements_by_xpath('//img[@class="lazyload lazy"]')
    
    i=0  
    while i < len(musinsa):    
        img_ele = musinsa[i]
        driver.execute_script("arguments[0].scrollIntoView();", img_ele)
        img = img_ele.get_attribute('src')
        
        if "0.gif" in img:
            time.sleep(1)          
            continue  #밑에 코드 무시하고 반복해라. 
            
        img_file = f"./img_mu/{page}_{i}.jpg"
        urllib.request.urlretrieve(img, img_file)
        ele = img_ele.find_element_by_xpath('./../../..')
        name = ele.find_element_by_xpath('./div/p[@class="list_info"]').text
        price = ele.find_element_by_xpath('./div/p[@class="price"]').text
        
        try :
            del_price = ele.find_element_by_xpath('./div/p[@class="price"]/del').text
            price = price.replace(del_price,"")
        except:
            pass        
        
        print(img,name,price)
        df.loc[len(df)] = [img_file, name, price]
        i += 1
        
```
